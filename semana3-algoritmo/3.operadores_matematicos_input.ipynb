```python
def calculadora(primeiro_numero, segundo_numero, operador_opcao):

    if operador_opcao == 1:
      resultado = primeiro_numero + segundo_numero
    elif operador_opcao == 2:
      resultado = primeiro_numero - segundo_numero
    elif operador_opcao == 3:
      resultado = primeiro_numero * segundo_numero
    elif operador_opcao == 4:
      resultado = primeiro_numero / segundo_numero
    else:
      resultado = 0
    return resultado

print("============Calculadora===========")
operacao = int(input("Escolha a operação que deseja realizar:\n1-Adição \n2-Subtração\n3-Multiplicação\n4-Divisão\n"))
numero1 = float(input("Digite o primeiro número:"))
numero2 = float(input("Digite o segundo número:"))
calc =  calculadora(numero1,numero2,operacao)

print(calc)
if calculadora == 0:
  print("OPeração inválida! Escolha uma das opções descritas!")
```
